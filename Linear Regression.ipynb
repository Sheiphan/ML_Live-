## Polynomial Regression

```python
df
```

```python
X1 = df.drop(['Sales', 'total_spend','Unnamed: 0'], axis =1,)
X1
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_conv = PolynomialFeatures(degree=2 , include_bias=False)
```

```python
poly_conv.fit(X1)
```

```python
poly_features = poly_conv.transform(X1)   # tranfer 3 feature to 9 feature to get 9 
                                           #parameters for degree 3.
```

```python
poly_features
```

```python
poly_features.shape
```

```python
X1.iloc[0]
```

```python
poly_features[0]
```

```python
from sklearn.model_selection import train_test_split
X_trainP, X_testP, y_trainP, y_testP = train_test_split(poly_features, y, test_size=0.33, random_state=101)

```

```python
X_trainP
```

```python
model1 = LinearRegression()
```

```python
model1.fit(X_trainP, y_trainP)
```

```python
test_pred_poly = model1.predict(X_testP)
test_pred_poly
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_testP, test_pred_poly)
MAE
```

```python
MSE = mean_squared_error(y_testP, test_pred_poly)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
model1.coef_
```

```python
# Checking all model from degrees 1 to 9
train_rmse_errors = []
test_rmse_errors = []

for d in range(1,10):
    poly_converter = PolynomialFeatures(degree=d, include_bias=False)
    poly_features = poly_converter.fit_transform(X1)

    X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.33, random_state=101)

    model = LinearRegression()
    model.fit(X_train, y_train)

    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test,test_pred))

    train_rmse_errors.append(train_rmse)
    test_rmse_errors.append(test_rmse)
```

```python
train_rmse_errors
```

```python
test_rmse_errors   # overfitting is happening after the 4th degree
```
